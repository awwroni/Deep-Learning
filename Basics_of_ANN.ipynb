```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers
```

In machine learning, we use standardization -- keeping the avrage around 0, so the vlaes range between -3 to 3

In deep learning, we use normalization, the vlaues are made to narrow down between 0 and 1  

For normalization, we use min max scaler

```python
df = pd.read_csv('plant_health_data.csv');
```

```python
X = df[['Soil_Moisture', 'Ambient_Temperature',
       'Soil_Temperature', 'Humidity', 'Light_Intensity', 'Soil_pH']]
```

```python
y = df['Plant_Health_Status']

```

```python
X_min = X.min()
X_max = X.max()

X_scaled = (X - X_min)/(X_max-X_min + 1e-8)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Re-split the data with encoded labels
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.25, random_state=42)

# Determine the number of unique classes
num_classes = len(le.classes_)

model = keras.Sequential([
    layers.Input(shape=(X_train.shape[1],)),
    layers.Dense(8, activation='relu'),
    layers.Dense(num_classes, activation='sigmoid') # Changed for multi-class classification
])
```

```python
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['accuracy']) # Changed loss function
```

```python
history = model.fit(
    X_train.values, y_train,
    validation_data=(X_test.values, y_test),
    epochs=100,
    batch_size=4, verbose=1
)
```
